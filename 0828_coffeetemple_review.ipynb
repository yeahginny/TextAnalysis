```python
import pandas as pd

# 데이터를 pandas DataFrame으로 불러옴
review_data = pd.read_excel("review_20230725_091330_da.xlsx")
```

```python
import sqlite3

# 1. SQLite3를 사용하여 데이터베이스 파일 생성
conn = sqlite3.connect("reviews.db")

# 2. 해당 데이터베이스에 테이블을 생성하고 데이터를 입력
review_data.to_sql('product_reviews', conn, if_exists='replace', index=False)

# 3. 데이터를 조회하여 확인
query = "SELECT * FROM product_reviews LIMIT 5"
result = pd.read_sql(query, conn)
result
```

```python
import re

# 1. 정규 표현식 패턴 일치 함수 정의
def regexp(expr, item):
    reg = re.compile(expr)
    return reg.search(item) is not None

# 2. 함수를 SQLite 연결에 REGEXP 연산자로 연결
conn.create_function("REGEXP", 2, regexp)

# 이제 REGEXP 연산자를 사용하여 쿼리를 수행할 수 있습니다.
query = """
  SELECT 리뷰상세내용
  FROM product_reviews
  WHERE 리뷰상세내용 REGEXP '커피'
"""
result_regexp = pd.read_sql(query, conn)
result_regexp.head()
```

```python
# 커피 맛별 구매자 평점 집계
"""
SELECT
    CASE
        WHEN 리뷰상세내용 LIKE '%쥬시%' THEN '쥬시'
        WHEN 리뷰상세내용 LIKE '%아메리카노%' THEN '아메리카노'
        WHEN 리뷰상세내용 LIKE '%에스프레소%' THEN '에스프레소'
        WHEN 리뷰상세내용 LIKE '%드립백%' THEN '드립백'
        WHEN 리뷰상세내용 LIKE '%캡슐%' THEN '캡슐'
        ELSE '기타'
    END AS 커피맛,
    AVG(구매자평점) AS 평균평점
FROM product_reviews
GROUP BY 커피맛;
"""

result = pd.read_sql(query, conn)
result
```

```python
# 커피 맛별 시간별 평점 집계
"""
SELECT SUBSTR(리뷰등록일, 16, 2) AS 시간,
    CASE
        WHEN 리뷰상세내용 LIKE '%쥬시%' THEN '쥬시'
        WHEN 리뷰상세내용 LIKE '%아메리카노%' THEN '아메리카노'
        WHEN 리뷰상세내용 LIKE '%에스프레소%' THEN '에스프레소'
        WHEN 리뷰상세내용 LIKE '%드립백%' THEN '드립백'
        WHEN 리뷰상세내용 LIKE '%캡슐%' THEN '캡슐'
        ELSE '기타'
    END AS 커피맛,
    AVG(구매자평점) AS 평균평점
FROM product_reviews
GROUP BY SUBSTR(리뷰등록일, 16, 2), 커피맛
order by SUBSTR(리뷰등록일, 16, 2);
"""

result = pd.read_sql(query, conn)
result
```

```python
#  리뷰 구분별 일별 평점 집계
"""
SELECT SUBSTR(리뷰등록일, 1, 10) AS 날짜,
  리뷰구분,
    AVG(구매자평점) AS 평균평점
FROM product_reviews
GROUP BY SUBSTR(리뷰등록일, 1, 10), 리뷰구분;
"""
result = pd.read_sql(query, conn)
result
```

```python
# 리뷰 구분별 시간별 평점 집계
"""
SELECT SUBSTR(리뷰등록일, 16, 2) AS 날짜,
  리뷰구분,
    AVG(구매자평점) AS 평균평점
FROM product_reviews
GROUP BY SUBSTR(리뷰등록일, 16, 2), 리뷰구분;
"""

result = pd.read_sql(query, conn)
result
```

```python
# 상품명에 '캡슐'이 포함되고, '맛'라는 맛이 들어간 리뷰 찾기
"""
SELECT 리뷰상세내용
FROM product_reviews
WHERE 상품명 LIKE '%캡슐%' AND 리뷰상세내용 LIKE '%맛%';
"""
```

```python
# 평점 4 이상의 상품 중에서 '에스프레소'라는 맛이 들어간 리뷰의 상품명과 평점을 찾기
"""
SELECT 상품명, 구매자평점
FROM product_reviews
WHERE 구매자평점 >= 4 AND 리뷰상세내용 LIKE '%에스프레소%';
"""
```

```python
# '캡슐' 상품 중 가장 많은 리뷰를 받은 상품명 찾기
"""
SELECT 상품명, COUNT(리뷰상세내용) AS 리뷰개수
FROM product_reviews
WHERE 상품명 LIKE '%캡슐%'
GROUP BY 상품명
ORDER BY 리뷰개수 DESC
LIMIT 1;
"""
```

```python
# 리뷰에 '드립백'이라는 맛이 들어간 상품 중 평점이 가장 높은 상품 찾기
"""
SELECT 상품명, MAX(구매자평점) AS 최고평점
FROM product_reviews
WHERE 리뷰상세내용 LIKE '%드립백%';
"""
```

```python
# '아메리카노'라는 맛을 포함한 리뷰 중 가장 오래된 리뷰 찾기
"""
SELECT 리뷰상세내용, SUBSTR(리뷰등록일, 1, 10) as 일자
FROM product_reviews
WHERE 리뷰상세내용 LIKE '%아메리카노%'
ORDER BY SUBSTR(리뷰등록일, 1, 10)
LIMIT 1;
"""
```
