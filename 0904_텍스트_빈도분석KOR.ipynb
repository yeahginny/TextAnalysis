```python
import json

mz_news = json.loads(open('MZ세대_naver_news.json', 'r', encoding = 'utf-8').read())
mz_news
```

```python
# 잘못된 예시
import re
content = ''
for item in mz_news:
    if 'content' in item.keys():
        content = content+re.sub(r'[^\w]',' ',item['content'])+''
content
```

```python
# 잘 된예시
import re
content = ''
if 'content' in mz_news.keys():
    content = re.sub(r'[^\w]', ' ', str(mz_news['content']))

content[:1000]
```

```python
from konlpy.tag import Okt

nlp = Okt()
content_N = nlp.nouns(content)
content_N
```

```python
from collections import Counter
```

```python
count = Counter(content_N)
count
```

```python
word_count = dict()
for tag, counts in count.most_common(80):
  if (len(str(tag))>1):
    word_count[tag] = counts
    print("%s : %d" %(tag,counts))
```
