```python
import pandas as pd

reviews_df = pd.read_excel('review_20230725_091330.xlsx')

reviews_df.head()
```

```python
review_texts = " ".join(reviews_df['리뷰상세내용'].dropna())
```

```python
review_texts
```

```python
from konlpy.tag import Okt

nlp = Okt()
content_review = nlp.nouns(review_texts)
content_review
```

```python
from collections import Counter
```

```python
count_review = Counter(content_review)
count_review
```

```python
word_count = dict()
for tag, counts in count_review.most_common(80):
  if (len(str(tag))>1):
    word_count[tag] = counts
    print("%s : %d" %(tag,counts))
```

```python
plt.figure(figsize = (12,5))
plt.xlabel('키워드')
plt.ylabel('빈도수')
plt.grid(True)
sorted_Keys = sorted(word_count, key = word_count.get, reverse = True)
sorted_Values = sorted(word_count.values(), reverse = True)
plt.bar(range(len(word_count)), sorted_Values, align='center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation='vertical')
plt.show()
```
